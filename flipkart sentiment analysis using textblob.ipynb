# Calculating polarity score and subjective scores for all reviews

```python
# code for get polarity score and subjectivity score  of review
polarity = []
subjectivity = []
for i in range(df.shape[0]):
  blob = TextBlob(df.iloc[i][5])
  polarity.append(blob.sentiment[0])
  subjectivity.append(blob.sentiment[1])
```

```python
print(polarity)
print(len(polarity))
```

```python
print(subjectivity)
print(len(subjectivity))
```

```python
# Adding polarity score column  to dataframe
df["Polarity_score"] = polarity
```

```python
# Adding Subjectivity score column  to dataframe
df["Subjectivity_score"] = subjectivity
```

```python
df.head()
```

# Adding custom columns for further analysis

```python
# creating list of review polarity type 
review_polarity_type = []
for i in df["Polarity_score"]:
  if i > 0:
    review_polarity_type.append("positive")
  elif i < 0:
    review_polarity_type.append("negative")
  else:
    review_polarity_type.append("neutral")
print(review_polarity_type)
print(len(review_polarity_type))   
```

```python
# creating review subjective type
review_subjective_type = []
for i in df["Subjectivity_score"]:
  if i > 0.5:
    review_subjective_type.append("fact")
  elif i < 0.5:
    review_subjective_type.append("opinion")
  else:
    review_subjective_type.append("neutral")
print(review_subjective_type)
print(len(review_subjective_type))
```

```python
# Adding these two column in dataframe
df["review_polarity_type"] = review_polarity_type
df["review_subjective_type"] = review_subjective_type
```

```python
# creating dataframe for counts of types in "review_polarity_type"
df1 = df[["review_polarity_type"]]
polarity_df = df1.groupby(df["review_polarity_type"], sort = True).count()
polarity_df = polarity_df.rename(columns={"review_polarity_type" : "counts"})
polarity_df
```

```python
# Creating barplot for count of type in "review_polarity_type"
plt.figure(figsize=(5,3))
plt.bar(polarity_df.index,polarity_df["counts"],color= "r",edgecolor="m",linewidth = 2 )
plt.title("Counts of review_polarity_type")
plt.xlabel("Types")
plt.ylabel("counts")
```
