```python
# importing required liabraries
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
base_url = "https://www.flipkart.com/lois-caron-lcs-8188-trending-day-date-functioning-boys-analog-watch-men/product-reviews/itmc2c03ccfbe758?pid=WATFN6WFBSBYHTCX&lid=LSTWATFN6WFBSBYHTCXHUNGJW&marketplace=FLIPKART"
```

```python
urls=["https://www.flipkart.com/lois-caron-lcs-8188-trending-day-date-functioning-boys-analog-watch-men/product-reviews/itmc2c03ccfbe758?pid=WATFN6WFBSBYHTCX&lid=LSTWATFN6WFBSBYHTCXHUNGJW&marketplace=FLIPKART"]
for i in range(2,41):
  urls.append(base_url+"&page="+str(i))
print(urls)
```

```python
len(urls)
```

```python
cus_name =[]
loc=[]
cus_rev =[]
cus_rate = []
rev_date =[]
for url in urls:
  # requesting website to scrape the data
  source = requests.get(url)

  # creating a object soup
  soup = BeautifulSoup(source.content,'html.parser')

  # customer name
  names = soup.find_all('p',class_="_2sc7ZR _2V5EHH _1QgsS5")
  customer_names =[]
  for i in names:
    customer_names.append(i.text)

  # location of customer
  place = soup.find_all('p',class_="_2mcZGG")
  location = []
  for i in place:
    location.append(i.find_all('span')[1].text.split(',')[1])

  # code for customer review
  review=soup.find_all('div',class_="_6K-7Co")
  customer_review = []
  for i in review:
    customer_review.append(i.text)

  # code to scrape the rating
  rating = soup.find_all('div',class_="t-ZTKy _1QgsS5")
  customer_rating= []
  for i in rating:
    customer_rating.append(i.text[0])


  # code to scrape customer review date
  date = soup.find_all('p',class_='_2sc7ZR')[1::2]
  cus_rev_date=[]
  for i in date:
    cus_rev_date.append(i.text)

  cus_name.extend(customer_names)
  loc.extend(location)
  cus_rev.extend(customer_review)
  cus_rate.extend(customer_rating)
  rev_date.extend(cus_rev_date)
  
  
```

```python
df = pd.DataFrame({"Name" : cus_name,
                   "Location": loc,
                   "Review_date" : rev_date,
                   "Rating" : cus_rate,
                   "Review" : cus_rev })
```
